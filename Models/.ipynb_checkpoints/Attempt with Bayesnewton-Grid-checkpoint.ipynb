```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import sys
sys.path.insert(1, '../Preprocessing/')
import utils
import matplotlib.cm as cm
import bayesnewton
import objax
```

```python
data =  pd.read_csv('../Data/pv_power_df_5day.csv', index_col='datetime')
uk_pv = pd.read_csv('../Data/system_metadata_location_rounded.csv')
data_multiple = data.iloc[:, :50][:5000]
data_multiple.plot()
lats = dict(uk_pv.set_index('ss_id')['latitude_rounded'])
longs = dict(uk_pv.set_index('ss_id')['longitude_rounded'])
a = data_multiple.reset_index()
```

```python
def datetime_to_epoch(datetime):
    """
        Converts a datetime to a number
        args:
            datatime: is a pandas column
    """
    return datetime.astype('int64') / 1e18
```

```python
stacked = pd.DataFrame()

for column in a.drop(columns=['datetime']).columns:
    df = a[['datetime', column]]
    df = df.assign(farm = int(df.columns[-1])).rename(columns = {column:'PV'})
    stacked = pd.concat([stacked, df])
    
stacked['latitude'] = stacked['farm'].map(lats)
stacked['longitude'] = stacked['farm'].map(longs)
stacked['datetime'] = pd.to_datetime(stacked['datetime'])
stacked['epoch'] = datetime_to_epoch(stacked['datetime'])
#create a dictionary from epoch to datetime
epoch_to_datetime_dict = dict(zip(stacked.epoch, stacked.datetime)) 

stacked.head()
```

```python
X = np.array(stacked[['epoch', 'longitude', 'latitude']])
X = np.vstack([X[:, 0],X[:, 1],X[:, 2]]).T
Y = np.array(stacked[['PV']])

grid = True
print(Y.shape)
print("num data points =", Y.shape[0])

#Get a test set which is random;y selected 10% of the dataset
test_ind = np.random.permutation(X.shape[0])[:X.shape[0]//10]
t_test = X[test_ind, :1]
R_test = X[test_ind, 1:]
Y_test = Y[test_ind, :]
```

```python
if grid:
    # the gridded approach:
    t, R, Y = bayesnewton.utils.create_spatiotemporal_grid(X, Y)
else:
    # the sequential approach:
    t = X[:, :1]
    R = X[:, 1:]
```

```python
plt.figure(figsize=(12,7))
plt.plot(stacked['datetime'].unique(), Y[:,0])
```

```python
Nt = t.shape[0]
print("num time steps =", Nt)
N = Y.shape[0] * Y.shape[1] * Y.shape[2]
print("num data points =", N)
```

```python
#Here we select what the size of the grid should be, at what frequency, and then create the meshgrid using linspace
#

N_test = 20  #50

# r1 = np.unique(X[:, 1])
# r2 = np.unique(X[:, 2])
X1range = max(X[:, 1]) - min(X[:, 1])
X2range = max(X[:, 2]) - min(X[:, 2])

r1 = np.linspace(min(X[:, 1]) - 0.1 * X1range, max(X[:, 1]) + 0.1 * X1range, num=N_test)
r2 = np.linspace(min(X[:, 2]) - 0.05 * X2range, max(X[:, 2]) + 0.05 * X2range, num=N_test)

rA, rB = np.meshgrid(r1, r2)
r = np.hstack((rA.reshape(-1, 1), rB.reshape(-1, 1)))  # Flattening grid for use in kernel functions
Rplot = np.tile(r, [t.shape[0], 1, 1])
```

```python
var_y = 0.5
var_f = 0.5
len_time = 5  # step size = 1 (hour)
len_space = 1  # spatial inputs normalised to around [-3, 3]
```

```python
sparse = False
opt_z = False  # will be set to False if sparse=False

if sparse:
    #NOTE: HERE Z ARE THE INDUCING POINTS OF THE SPARSE APPROXIMATION
    z1 = np.linspace(np.min(X[:, 1]), np.max(X[:, 1]), num=7)
    z2 = np.linspace(np.min(X[:, 2]), np.max(X[:, 2]), num=7)
    zA, zB = np.meshgrid(z1, z2)  # Adding additional dimension to inducing points grid
    z = np.hstack((zA.reshape(-1, 1), zB.reshape(-1, 1)))  # Flattening grid for use in kernel functions
    del z1, z2, zA, zB
else:
    z = R[0, ...]
```

```python
len_day = stacked[stacked['datetime'] == '2018-01-01 16:00:00'].iloc[0]['epoch'] - stacked[stacked['datetime'] == '2018-01-01 08:00:00'].iloc[0]['epoch']
len_year = len_day * 365.25

#FOLLOWING THE EXAMPLE OF PAPER 3 CHAPTER 4.3 + THE DEMO FROM THIS NOTEBOOK

kern_time_day = bayesnewton.kernels.QuasiPeriodicMatern32(variance=var_f,
                                                    lengthscale_periodic=len_day,
                                                    period=len_day,
                                                    lengthscale_matern=len_day,
                                                    order=6) #ORDER IS THE ORDER OF THE APPROXIMATION OF THE PERIODIC KERNEL TO TRANSFORM INTO STATE SPACE
kern_time_year = bayesnewton.kernels.QuasiPeriodicMatern32(variance=var_f,
                                                    lengthscale_periodic=len_year,
                                                    period=len_year,
                                                    lengthscale_matern=len_year,
                                                    order=6) #ORDER IS THE ORDER OF THE APPROXIMATION OF THE PERIODIC KERNEL TO TRANSFORM INTO STATE SPACE



kern_space0 = bayesnewton.kernels.Matern32(variance=var_f, lengthscale=len_space)

kern_space1 = bayesnewton.kernels.Matern32(variance=var_f, lengthscale=len_space)

kern_space = bayesnewton.kernels.Separable([kern_space0, kern_space1])

```

```python
kern = bayesnewton.kernels.SpatioTemporalKernel(temporal_kernel=kern_time,
                                                spatial_kernel=kern_space,
                                                z=z,
                                                sparse=sparse,
                                                opt_z=opt_z,
                                                conditional='Full')
```

```python
lik = bayesnewton.likelihoods.Gaussian(variance=var_y)
# model = bayesnewton.models.VariationalGP(kernel=kern, likelihood=lik, X=X, Y=y)

#ESSENTIALLY, THIS MODEL CONSIDER THE X DIMENSIONS MARKOVIAN,WITH Z INDUCING POINTS, AND THUS
#OBTAINS A RECURRENT KALMAN-LIKE ALGORITHM IN THE SPACE-TIME DIMENSION
# http://proceedings.mlr.press/v130/wilkinson21a/wilkinson21a.pdf
model = bayesnewton.models.MarkovVariationalGP(kernel=kern, likelihood=lik, X=t, R=R, Y=Y)
```

```python
lr_adam = 0.05
lr_newton = 0.5
iters = 10
opt_hypers = objax.optimizer.Adam(model.vars())
energy = objax.GradValues(model.energy, model.vars())
```

```python
@objax.Function.with_vars(model.vars() + opt_hypers.vars())
def train_op():
    model.inference(lr=lr_newton)  # perform inference and update variational params
    dE, E = energy()  # compute energy and its gradients w.r.t. hypers
    opt_hypers(lr_adam, dE)
    return E
```

```python
train_op = objax.Jit(train_op)

t0 = time.time()
for i in range(1, iters + 1):
    loss = train_op()
    print('iter %2d, energy: %1.4f' % (i, loss[0]))
t1 = time.time()
print('optimisation time: %2.2f secs' % (t1-t0))
```
